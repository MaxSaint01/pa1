```python
P = wages.loc[wages["Full or Part-Time"] == "P"]
chart = P["Department"].unique()
darn = []
logs = []
for i in chart:
    darn.append(i)
    logs.append(wages.loc[wages["Department"] == i].shape[0])
goob = pd.DataFrame({"Department": darn})
Count = []
for k in goob["Department"]:
    Count.append(P.loc[P["Department"]==k].shape[0])
goob["Count"] = Count
goob["Total"] = logs
goob["Ratio"] = goob["Count"]/goob["Total"]*100
dog = goob.sort_values(by=["Count"], ascending = True)
dork = goob.sort_values(by=["Ratio"], ascending = True)
```
