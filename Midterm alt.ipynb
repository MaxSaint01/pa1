```python
fig, ax1 = plt.subplots(figsize=(15, 8))
ax2 = ax1.twinx()

ax1.plot(us2k.year, us2k.population1517 / 10**6, '.r-')
ax1.plot(us2k.year, us2k.population1819 / 10**6, '.r-')
ax2.plot(us2k.year, -us2k.abortionstotal/ 10**6, '.b-')

ax1.set_xlabel('Year',fontsize = 20)
ax1.set_ylabel('US population (15-19 years old) (millions)', color='g',fontsize = 20)
ax2.set_ylabel('Number of US abortions (millions)', color='b',fontsize = 20)
ax1.set_title('Abortion has been killing our teenagers over the past decade',
             fontsize = 20)
ax1.tick_params(axis='both', which='major', labelsize=15)
ax2.tick_params(axis='both', which='major', labelsize=15)
ax2.tick_params(axis='both', which='minor', labelsize=15)
plt.show()
```
